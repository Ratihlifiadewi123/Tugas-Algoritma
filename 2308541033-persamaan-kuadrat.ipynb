```python
import numpy as np
import matplotlib.pyplot as plt
import cmath as cmath

def quadratic_function(x_33, a_33, b_33, c_33):
    return a_33 * x_33**2 + b_33 * x_33 + c_33

# Menentukan parameter fungsi kuadrat
a_33 = float(input("a adalah "))
b_33 = float(input("b adalah "))
c_33 = float(input("c adalah "))

# Membuat kumpulan nilai x dari -10 hingga 10
x_values_33 = np.linspace(-10, 10, 400)

# Menghitung nilai y untuk setiap x
y_values_33 = quadratic_function(x_values_33, a_33, b_33, c_33)

# Menggambar grafik fungsi kuadrat
plt.plot(x_values_33, y_values_33, label=f'y = {a_33}x^2 + {b_33}x + {c_33}')
plt.xlabel('x_33')
plt.ylabel('y_33')
plt.title('Grafik Fungsi Kuadrat')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)

# Menentukan titik puncak dan sumbu simetris
h_33 = -b_33/(2*a_33)
k_33 = quadratic_function(h_33, a_33, b_33, c_33)
plt.scatter(h_33, k_33, color='red', label='Titik Puncak')
plt.axvline(h_33, color='green', linestyle='--', label='Sumbu Simetris')

#bentuk kurva
if a_33 > 0:
    print ("kurva terbuka keatas")
else:
    print ("kurva terbuka kebawah") 

# Menentukan posisi titik puncak
if h_33 > 0:
    posisi_33 = 'di kanan sumbu y'
elif h_33 == 0:
    posisi_33 = 'pada sumbu y'
else:
    posisi_33 = 'di kiri sumbu y'
plt.text(h_33, k_33, f'({h_33},{k_33})\n{posisi_33}', ha='right')
    
# Bagaimana kondisi parabola memotong sumbu y
if c_33 > 0:
    potong_sumbu_y_33 = 'positif'
elif c_33 == 0:
    potong_sumbu_y_33 = 'di pusat (0,0)'
else:
    potong_sumbu_y_33 = 'negatif'
plt.text(0, c_33, f'Potong sumbu y {potong_sumbu_y_33}', ha='left')

plt.legend()
plt.show()

# menentukan diskriminan
D_33 = float(b_33**2-4*a_33*c_33)
print ("Nilai diskriminan adalah D={}". format(D_33))
if D_33 >= 0:
    print ("Mempunyai akar-akar persamaan")
    x_1_33 = (-b_33+cmath.sqrt(D_33))/(2*a_33)
    x_2_33 = (-b_33-cmath.sqrt(D_33))/(2*a_33)
    print (x_1_33,x_2_33)
else:
    print ("Akar-akar persamaan imajiner")
    
#bentuk persamaan garis
x_1_33 = (-b_33+cmath.sqrt(D_33))/(2*a_33)
x_2_33 = (-b_33-cmath.sqrt(D_33))/(2*a_33)
print ("Persamaan kuadrat adalah x^2 + {}x + {} = 0". format(x_1_33,x_2_33))
```
