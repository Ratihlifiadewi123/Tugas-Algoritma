```python
jam = float(input("Jam : "))
waktu = input("Ubah jam dalam menit atau detik? ")
if waktu.upper() == 'MENIT':
    menit = jam*60
    print(str(jam) + " jam yaitu " + str(menit) + " menit")
if waktu.upper() == 'DETIK':
    detik = jam*3600
    print(str(jam) + " jam yaitu " + str(detik) + " detik")
```
