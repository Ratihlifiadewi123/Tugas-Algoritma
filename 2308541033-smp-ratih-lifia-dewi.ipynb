```python
#Mencari nilai dari n dengan menggunakan sifat logaritma 

import cmath as cmath
import math as math

#Diketahui nilai a, b, dan c
a_33 = 1
b_33 = 1
c_33 = -1

#Mencari x1 dan x2
x1_33 = (-(b_33) + cmath.sqrt(b_33**2-4*a_33*c_33))/ 2*a_33
x2_33 = (-(b_33) - cmath.sqrt(b_33**2-4*a_33*c_33))/ 2*a_33
print("Akar pertama yaitu {}".format(x1_33.real))
print("Akar kedua yaitu {}".format(x2_33.real))

#Mencari nilai n dengan memasukkan salah satu akar persamaan yaitu akar pertama atau akar kedua
n_33 =(input("Mencari nilai n dengan memasukkan salah satu akar (x1 atau x2) ke persamaan n: "))

#Memasukan nilai x untuk menentukan nilai n
if n_33 == "x1":
    n1_33 = (2024*math.log(abs(x1_33)) + math.log(abs(1-x1_33)))/ (math.log(abs(x1_33)) + math.log(abs(1-x1_33)))
    print ("Nilai n yang memenuhi persamaan adalah {}".format(n1_33))
elif n_33 == "x2":
    n2_33 = (2024*math.log(abs(x2_33)) + math.log(abs(1-x2_33)))/ (math.log(abs(x2_33)) + math.log(abs(1-x2_33)))
    print ("Nilai n yang memenuhi persamaa adalah {}".format(n2_33))
else:
    print ("Masukkan salah satu akar yaitu x1 atau x2")
```
