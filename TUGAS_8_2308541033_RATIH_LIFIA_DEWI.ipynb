```python
import numpy as np
import matplotlib.pyplot as plt

# Persamaan garis yang melewati 2 titik

# Garis pertama melewati (o_33,p_33)an (q_33,r_33)
o_33 = float(input("Nilai x dari titik pertama dari garis pertama adalah "))
p_33 = float(input("Nilai y dari titik pertama dari garis pertama adalah "))
q_33 = float(input("Nilai x dari titik kedua dari garis pertama adalah "))
r_33 = float(input("Nilai y dari titik kedua dari garis pertama adalah "))

# Garis kedua melewati (i_33,j_33) dan (k_33,l_33)
i_33 = float(input("Nilai x dari titik pertama dari garis kedua adalah "))
j_33 = float(input("Nilai y dari titik pertama dari garis kedua adalah "))
k_33 = float(input("Nilai x dari titik kedua dari garis kedua adalah "))
l_33 = float(input("Nilai y dari titik kedua dari garis kedua adalah "))

m_1_33 = (r_33-p_33)/(q_33-o_33) # Gradien garis pertama
m_2_33 = (l_33-j_33)/(k_33-i_33) # Gradien garis kedua

# Persamaan garis pertama adalah y_33=m_1_33(x_33-o_33)+p_33=m_1_33x_33-m_1_33a_33+b_33
# Persamaan garis kedua adalah y_33=m_2_33(x_33-e_33)+f_33=m_2_33x_33-m_2_33e_33+f_33

c_1_33 = -m_1_33*o_33+p_33 # Titik potong garis pertama terhadap sumbu y
c_2_33 = -m_2_33*i_33+j_33 # titik potong garis kedua terhadap sumbu y garis kedua

print("Persamaan garis pertama adalah y={}x+{}".format(m_1_33,c_1_33))
print("Persamaan garis kedua adalah y={}x+{}".format(m_2_33,c_2_33))

# Terdapat 3 kemungkinan dari kedua persamaan garis ini:
if m_1_33==m_2_33:
    #print("Kedua garis sejajar sehingga kemungkinannya adalah tidak terdapat solusi atau terdapat tak berhingga banyaknya solusi")
# 1. Terdapat tak berhingga banyaknya solusi (kedua garis berhimpitan) (kedua garis sejajar)
    if c_1_33 == c_2_33:
        print("Kedua garis berhimpitan sehingga akan terdapat tak berhingga banyaknya solusi")
        x_33=np.linspace(-5,5,100)
        plt.plot(x_33,m_1_33*(x_33-o_33)+p_33,color='green')
        plt.plot(x_33,m_2_33*(x_33-i_33)+j_33,color='purple')
        plt.show()
        
# 3. Tidak terdapat perpotongan garis (tidak terdapat solusi)
    else:
        print("Kedua garis tidak mungkin berpotongan, sehingga tidak akan terdapat solusi.")
        x_33=np.linspace(-5,5,100)
        plt.plot(x_33,m_1_33*(x_33-o_33)+p_33,color='green')
        plt.plot(x_33,m_2_33*(x_33-i_33)+j_33,color='purple')
        plt.show()

# 3. Terdapat perpotongan di satu titik (gradien kedua garis tidak sama) 
else:
    print("Akan terdapat satu titik potong")
    tp_x_33 = (-m_2_33*i_33+j_33+m_1_33*o_33-p_33)/(m_1_33-m_2_33)
    tp_y_33 = m_1_33*(tp_x_33-o_33)+p_33
    print("Titik potong kedua garis adalah ({},{})".format(tp_x_33,tp_y_33))
    
    x=np.linspace(tp_x_33-5,tp_x_33+5,100)
    plt.plot(x,m_1_33*(x_33-o_33)+p_33,color='green')
    plt.plot(x,m_2_33*(x_33-i_33)+j_33,color='purple')
    plt.axvline(x=0,color='black')
    plt.axhline(y=0,color='black')
    plt.show()
    
```
